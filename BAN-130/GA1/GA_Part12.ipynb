Part 1: Exploring and Filtering Data (10%)
1. Load the EU_OCC.sas7bdat file and examine the column names and values.
2. Filter the dataset to include only rows where YearMon starts with "2016".
3. Format the columns Hotel, ShortStay, and Camp to include commas as thousand
separators.
4. Drop the column Geo from the dataset.
Deliverables: Python script that completes the above tasks, with clear comments

```python
# 1.Load the EU_OCC.sas7bdat file and examine the column names and values.
import pandas as pd
df_EU_OCC = pd.read_sas('https://raw.githubusercontent.com/amknm2026/BAN2026/main/BAN-130/GA1/Data/eu_occ.sas7bdat', format='sas7bdat',encoding="cp1252")

print ('1. Load EU_OCC.sas7bdat file')
print ('Display Sample for sanity check');
display (df_EU_OCC.head(5))
print ()

# format column name and data type information for easier view
dtypes_table = (
  df_EU_OCC.dtypes
  .reset_index()
  .rename(columns={"index":"Column Name",0:"Data Type"})
  )
print ('Column Names and data Type information \n')
display (dtypes_table)


# there are no numbers, so no need to describe the dataset
print ('\n All variables are character, there are no summary info (mean, mediam) \n')

# format column name and number of empty occurrencies for easier view
dvempty_table = (
    df_EU_OCC.isna().sum()
    .reset_index()
    .rename(columns={"index":"Column Name",0:"Empty Values"})
)

print ('\n Total Missing Values \n')
display (dvempty_table);
print ()

#identify missing values for Hotel and select 5 records (sample)
df_HotelNA = df_EU_OCC[df_EU_OCC.Hotel.isna()].sample(5)

#identify missing values for ShotStay
df_SSNA = df_EU_OCC[df_EU_OCC.ShortStay.isna()]
#remove any matches with df_HotelNA and select 5 records (sample)
df_SSNA = pd.concat([df_SSNA,df_HotelNA]).drop_duplicates(keep=False).sample(5)

#identify missing values for Camp
df_CampNA = df_EU_OCC[df_EU_OCC.Camp.isna()]
#remove any matches with df_HotelNA and ShortStay
df_CampNA = pd.concat([df_CampNA,df_HotelNA, df_SSNA]).drop_duplicates(keep=False).sample(5)

len(df_CampNA)
print ('\n Missing values - Hotel -Sample \n ')
display(df_HotelNA)

print ('\n Missing values - ShortStay -Sample \n')
display(df_SSNA)

print ('\n Missing values - Camp - Sample \n')
display(df_CampNA)

```

```python
# 2.Filter the dataset to include only rows where YearMon starts with "2016".
df_EU_OCC_2016 = df_EU_OCC[df_EU_OCC.YearMon.str.startswith('2016')]

print ('2. Filter the dataset to include only rows where YearMon starts with "2016"')
print(f'Number of records: ',len(df_EU_OCC_2016) )
print('Sample - 5 records \n')
display(df_EU_OCC_2016.sample(5));
```

```python
# 3.Format the columns Hotel, ShortStay, and Camp to include commas as thousand separators.
# defined a named function rather than using a lambda (since we are beginners)
def add_thousand_separator(value):
  if pd.notna(value):
    formatted_value = '{:,}'.format(value)
    return formatted_value
  else:
    return value

#   Data conversion has not been requested, only keep character variables
#   To prevent conversion issues (if required converting to numeric) , a new dataset will be created to store these numbers
df_formatted_values = df_EU_OCC_2016.copy();


# select the columns required
cols =  ["Hotel", "ShortStay", "Camp"]
# applies the change to the 3 columns
for column_name in cols:
  df_formatted_values[column_name]= df_formatted_values[column_name].map(add_thousand_separator)

print ('3. Records from 2016 - Hotel, Shortsay and Camp columns including thousands separators')
print (f'Total Records from 2016 (sanity check): {len(df_formatted_values)} \n')
display(df_formatted_values.sample(5))

print ()
# 4.Drop the column Geo from the dataset. Deliverables: Python script that completes the above tasks, with clear comments
# will continue with previous formatted dataset (dt_formatted_values )
df_formatted_values.drop(columns=['Geo'], inplace=True)
print ('4. Records from 2016 - Hotel, Shortsay and Camp columns including thousands separators, Geo column removed')
print (f'Total Records (sanity check): {len(df_formatted_values)} \n')
display(df_formatted_values.sample(5))
```

Part 2: Creating and Formatting Columns (15%)
1. Load the np_summary.sas7bdat file.
2. Create a new column SqMiles by converting Acres to square miles (Acres * 0.0015625).
3. Create a new column Camping as the sum of OtherCamping, TentCampers, RVCampers,
and BackcountryCampers.
4. Format SqMiles and Camping columns to include commas and zero decimal places.
5. Save the updated dataset with only the following columns: Reg, ParkName, DayVisits,
OtherLodging, Acres, SqMiles, Camping.
Deliverables: Python script with the processed DataFrame.

```python
#1. Load the np_summary.sas7bdat file.

import pandas as pd
df_np_summary = pd.read_sas('https://raw.githubusercontent.com/amknm2026/BAN2026/main/BAN-130/GA1/Data/np_summary.sas7bdat', format='sas7bdat',encoding="cp1252")

print ('Load np_summary data file')
print (f'Number of Rows: {df_np_summary.shape[0]} ')
print (f'Number of Columns: {df_np_summary.shape[1]}\n' )


print ('Sample for sanity checking\n')
display(df_np_summary.sample(5))


#verify column information
# format column name and data type information for easier view
df_np_summary_types = (
  df_np_summary.dtypes
  .reset_index()
  .rename(columns={"index":"Column Name",0:"Data Type"})
  )

print ('Column Names and data Type information')
display (df_np_summary_types)
print ()

# format column name and number of empty occurrencies for easier view
df_np_summary_vempty = (
      df_np_summary.isna().sum()
      .reset_index()
      .rename(columns={"index":"Column Name",0:"Empty Values"})
    )

print ('\n Missing Values\n')
display (df_np_summary_vempty);

print ('\n Describe Numeric values\n')
display(df_np_summary.describe())
print ()
```

```python
#2. Create a new column SqMiles by converting Acres to square miles (Acres * 0.0015625).

df_np_summary_new = df_np_summary.copy()

# use number indicated
multiplier =  0.0015625
# create new field in dataframe
df_np_summary_new['SqMiles'] = df_np_summary_new.Acres * multiplier

print ('New Column SqMiles Added \n')
display(df_np_summary_new.sample(5))

```

```python
#3.Create a new column Camping as the sum of OtherCamping, TentCampers, RVCampers, and BackcountryCampers.

#create new column adding values from All Campers
df_np_summary_new['Camping'] = df_np_summary_new.OtherCamping + df_np_summary_new.TentCampers+ df_np_summary_new.RVCampers + df_np_summary_new.BackcountryCampers

#display values
print ('New Column Camping Added \n')
display(df_np_summary_new.sample(5))
```

```python
#4.Format SqMiles and Camping columns to include commas and zero decimal places.

#creating copies for formatting
df_np_summary_fmt = df_np_summary_new.copy()

#to allow display with comma and decimal places the variables will be converted to character
df_np_summary_fmt["SqMiles"] = df_np_summary_fmt["SqMiles"].map("{:,.2f}".format)
df_np_summary_fmt.Camping = df_np_summary_fmt["Camping"].map("{:,.2f}".format)

print('New variables (SqMiles and Camping) now are formatted including zero and decimal places')
print('Sample is larger as an atempt to find large numbers')
display(df_np_summary_fmt.sample(15))
```

```python
#5.Save the updated dataset with only the following columns:
#Reg, ParkName, DayVisits, OtherLodging, Acres, SqMiles, Camping. Deliverables: Python script with the processed DataFrame.

df_np_summary_fmt.to_csv("np_summary_updated.csv",
                         columns=['Reg', 'ParkName', 'DayVisits', 'OtherLodging', 'Acres', 'SqMiles', 'Camping'],
                         index=False)

```
